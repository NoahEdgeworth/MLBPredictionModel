```python
import pandas as pd
```

```python
df = pd.read_csv('Data/processed/clean_batting_stats.csv')
df = df[df['PA'] >= 502] #502 is considered full time every day players
df['K%'] = df['K%'] * 100
```

```python
strikeouts = df.groupby(['Season', 'Name'])[['SO', 'HR', 'K%', 'PA']].sum()
strikeouts.reset_index()
strikeouts
```

```python
strikeouts = strikeouts.sort_values('SO', ascending=False)
```
