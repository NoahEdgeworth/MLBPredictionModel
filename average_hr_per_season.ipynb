```python
import pandas as pd
```

```python
df = pd.read_csv('Data/processed/clean_batting_stats.csv')
df
```

```python
full_time_player = df[df['PA'] >= 502] #502 is what MLB considers a full time every day player
full_time_player
```

```python
avg_hr_by_team_by_season = full_time_player.groupby(['Season', 'Team'])[['HR']].mean()
avg_hr_by_team_by_season
```
